```python
import pandas as pd
```

```python
# user file

u_columns = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_csv('../ml-100k/u.user', sep='|', 
                    names = u_columns, encoding='latin-1')

users.head()
```

```python
n_user = users.shape[0]
print('number of user:', n_user)
```

```python
# rating file

r_columns = ['user_id', 'movie_id', 'rating', 'timestamp']

ratings_base = pd.read_csv('../ml-100k/ua.base', sep = '\t', 
                           names = r_columns, encoding='latin-1')
ratings_test = pd.read_csv('../ml-100k/ua.test', sep = '\t', 
                           names = r_columns, encoding='latin-1')

ratings_base.head()
```

```python
ratings_test.head()
```

```python
import numpy as np

rate_train = np.array(ratings_base)
rate_test = np.array(ratings_test)

print(rate_train.shape, rate_test.shape)
```

```python
# film_type
items_columns = [
    'movie_id', 'movie_title', 'release_date', 'video_release_date',
    'IMDb_URL', 'unknown',
    'Action', 'Adventure',
    'Animation', 'Children\'s', 
    'Comedy', 'Crime', 'Documentary', 
    'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 
    'Musical', 'Mystery', 
    'Romance', 'Sci-Fi', 
    'Thriller', 'War', 'Western'
]

items = pd.read_csv('../ml-100k/u.item', sep = '|', 
                    names=items_columns, encoding='latin-1')

items.head()
```

```python
items_mat = np.array(items)
n_items = items_mat.shape[0]
print(n_items)
```

```python
X = items_mat[:, 6:]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

transformer = TfidfTransformer(smooth_idf=True, norm='l2')
tfidf = transformer.fit_transform(X.tolist())

print(tfidf)
```

```python
def get_rated_items(rate_matrix, user_id):
    all_user = rate_matrix[:,0]
    ids = np.where(all_user==user_id+1)[0]
    item_ids = rate_matrix[ids, 1] - 1
    scores = rate_matrix[ids, 2]
    
    return (item_ids, scores)
```

```python
from sklearn.linear_model import Ridge # ridge regression
from sklearn import linear_model

d = tfidf.shape[1]
W = np.zeros((d, n_user))
b = np.zeros((1, n_user))

for n in range(n_user):
    ids, scores = get_rated_items(rate_train, n)
    clf = Ridge(alpha = 0.01, fit_intercept = True)
    Xhat = tfidf[ids, :]
    
    clf.fit(Xhat, scores)
    W[:, n] = clf.coef_
    b[0, n] = clf.intercept_
```

```python
    Yhat = tfidf.dot(W) + b
```

```python
n = 10
np.set_printoptions(precision=2) # 2 digits after . 
ids, scores = get_rated_items(rate_test, n)
Yhat[n, ids]
print('Rated movies ids :', ids )
print('True ratings     :', scores)
print('Predicted ratings:', Yhat[ids, n])
```
